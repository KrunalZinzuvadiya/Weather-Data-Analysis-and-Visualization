```python
#What is the mean temperature recorded by month?
mean_temperatures = {}

for month in range(1, 13):
    # Filter for the current month and calculate the mean temperature
    mean_temperatures[month] = df.loc[df.index.month == month, 'Temp (C)'].mean()

# Convert the dictionary to a Pandas Series for better visualization
mean_temperatures_series = pd.Series(mean_temperatures)

print(mean_temperatures_series)
```

```python
import numpy as np
#or using pivot table
mean_temperature_df = df.pivot_table(values = 'Temp (C)', index = df.index.month, aggfunc = np.mean)
mean_temperature_df
```

```python
import calendar

def checkList(x):
    """Checks if a weather description contains both 'rain' and 'thunderstorm'."""
    weatherList = ["rain", "thunderstorm"]
    count = 0
    for substring in weatherList:
        if substring.lower() in x.lower():
            count += 1
    return count == 2

# Assuming your DataFrame is named 'df' with columns 'Date' and 'Weather'
df['Month'] = df.index.month
df['Month'] = df['Month'].apply(lambda x: calendar.month_name[x])

rained = df['Weather'].str.lower().apply(checkList)

print("Months in which it rained with thunderstorm:")
print(df[rained]['Month'].unique())
```

```python
# Count days with any precipitation
precipitation_conditions = ['Rain', 'Snow', 'Drizzle']
precipitation_days = df[df['Weather'].str.contains('|'.join(precipitation_conditions), na=False)]
total_precipitation_days = precipitation_days.resample('D').size()
print("Total Precipitation Days:\n", total_precipitation_days)
```

```python
# Calculate average wind speed by month
monthly_avg_wind_speed = df['Wind Spd (km/h)'].resample('M').mean()
print("Monthly Average Wind Speed:\n", monthly_avg_wind_speed)
```

```python
# Count occurrences of each weather condition
weather_counts = df['Weather'].value_counts()
print("Most Common Weather Conditions:\n", weather_counts.head(10))
```

```python
import matplotlib.pyplot as plt

# Plotting daily average temperatures over the year
daily_avg_temp = df['Temp (C)'].resample('D').mean()
plt.figure(figsize=(12, 6))
daily_avg_temp.plot(title='Daily Average Temperature in 2012', xlabel='Date', ylabel='Temperature (Â°C)')
plt.grid()
plt.show()
```

```python
# Find maximum and minimum temperatures recorded
max_temp = df['Temp (C)'].max()
min_temp = df['Temp (C)'].min()
print(f"Maximum Temperature: {max_temp} Â°C")
print(f"Minimum Temperature: {min_temp} Â°C")
```

```python
# Define seasons based on months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df.index.month.map(get_season)

# Calculate average temperature by season
seasonal_avg_temp = df.groupby('Season')['Temp (C)'].mean()
print("Average Temperature by Season:\n", seasonal_avg_temp)
```

```python
# Plotting the histogram of relative humidity
plt.figure(figsize=(10, 5))
plt.hist(df['Rel Hum (%)'], bins=20, color='green', alpha=0.7)
plt.title('Distribution of Relative Humidity in 2012')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Frequency')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Count occurrences of each weather condition
weather_counts = df['Weather'].value_counts()

# Plotting the bar chart for weather conditions
plt.figure(figsize=(12, 6))
weather_counts.plot(kind='bar', color='orange')
plt.title('Weather Conditions Count in 2012')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
